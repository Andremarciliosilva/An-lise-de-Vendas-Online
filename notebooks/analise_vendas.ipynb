```python
import pandas as pd
import matplotlib.pyplot as plt

# Ler o CSV
df = pd.read_csv('../output/sales_by_state.csv')

# Ordenar por vendas para deixar o gráfico mais informativo visualmente
df = df.sort_values(by="vendas_totais", ascending=False)

# Criar gráfico
plt.figure(figsize=(10, 6))
bars = plt.bar(
    df["estado"],
    df["vendas_totais"],
    color=plt.cm.viridis(df["vendas_totais"] / max(df["vendas_totais"]))
)

# Título e rótulos
plt.title("Vendas Totais por Estado", fontsize=16, weight='bold')
plt.xlabel("Estado", fontsize=12)
plt.ylabel("Vendas Totais", fontsize=12)

# Adicionar os valores no topo das barras
for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        yval + 2000,
        f'{yval:,}',  # Formata com vírgula como separador de milhar
        ha='center',
        va='bottom',
        fontsize=10
    )

# Grade e layout
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
# Ler o CSV
df = pd.read_csv('../output/best_selling_product.csv')

# Ordenar os dados para melhor visualização
df = df.sort_values(by="quantidade", ascending=False)

# Criar gráfico de pizza
plt.figure(figsize=(8, 8))
colors = plt.cm.Set3(range(len(df)))  # Usando Set3 para cores mais vivas

plt.pie(
    df["quantidade"],
    labels=df["categoria"],
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    wedgeprops={"edgecolor": "white", "linewidth": 1}
)

# Título e layout
plt.title("Produtos mais vendidos", fontsize=16, weight='bold')
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python

# Ler o CSV
df = pd.read_csv('../output/customers_most_purchases.csv')

# Criar gráfico de barras horizontais
plt.figure(figsize=(8, 5))
bars = plt.barh(
    df["cliente_id"].astype(str),  # Converte ID para string para melhor visualização
    df["quantidade"],
    color=plt.cm.coolwarm(df["quantidade"] / max(df["quantidade"]))  # Paleta suave e gradiente
)

# Título e rótulos
plt.title("Quantidade de Compras por Cliente", fontsize=16, weight='bold')
plt.xlabel("Quantidade", fontsize=12)
plt.ylabel("ID do Cliente", fontsize=12)

# Adicionar valores nas barras
for bar in bars:
    plt.text(
        bar.get_width() + 0.5,
        bar.get_y() + bar.get_height() / 2,
        f'{int(bar.get_width())}',
        va='center',
        fontsize=10
    )

# Estilo visual
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
# Ler o CSV
df = pd.read_csv('../output/monthly_recipe.csv')

# Converter coluna 'mes' para datetime
df["mes"] = pd.to_datetime(df["mes"])

# Criar gráfico de linha
plt.figure(figsize=(10, 5))
plt.plot(df["mes"], df["preco_unitario"], marker='o', linewidth=2.5, color='teal')

# Título e rótulos
plt.title("Receita mensal ao longo de 2023", fontsize=16, weight='bold')
plt.xlabel("Mês", fontsize=12)
plt.ylabel("Preço Unitário", fontsize=12)

# Estilo visual
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()

# Exibir o gráfico
plt.show()
```
