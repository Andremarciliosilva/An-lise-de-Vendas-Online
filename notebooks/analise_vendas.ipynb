```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Ler o CSV
df = pd.read_csv('../output/sales_by_state.csv')

# Criar gráfico de barras
plt.figure(figsize=(7, 3))
plt.bar(df['estado'], df['vendas_totais'], color='blue')

# Títulos e rótulos
plt.title('Sales by State', fontsize=12, fontweight='bold')
plt.xlabel('Estado', fontsize=4)
plt.ylabel('Vendas Totais (R$)', fontsize=4)

# Mostrar o gráfico
plt.show()
```

```python
# Ler o CSV
df = pd.read_csv('../output/best_selling_product.csv')

# Criar gráfico de barras
plt.figure(figsize=(8, 3))
plt.bar(df['categoria'], df['quantidade'], color='gray')

# Títulos e rótulos
plt.title('Best Selling Product', fontsize=14, fontweight='bold')
plt.xlabel('Categoria', fontsize=4)
plt.ylabel('Quantidade', fontsize=4)

# Mostrar o gráfico
plt.show()
```
